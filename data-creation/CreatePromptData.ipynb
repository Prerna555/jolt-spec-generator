```python
# Define a function to create the prompt
def create_prompt(row):
    system_prompt = "You are a JSON data mapper assistant that generates accurate spec files using Jolt.\n"
    user_prompt = f"Can you generate a Jolt spec to convert input json\n{row['input']}\nto output json\n{row['output']}\n"
    model_answer = f"spec {row['spec']}"

    return f"<s>[INST]<<sys>>{system_prompt}<</sys>>{user_prompt}[/INST]{model_answer}</s>"

# Apply the function to each row and create a new column 'formatted_prompt'
df['formatted_prompt'] = df.apply(create_prompt, axis=1)

# Save the reformatted data to a new Excel file
df=df[['formatted_prompt']]
```
